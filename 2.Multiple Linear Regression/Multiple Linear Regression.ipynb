## Importing the needed packages

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

## Importing the dataset

```python
df = pd.read_csv("50_Startups.csv")  # Load the dataset
df.head(9)  # Preview the first few rows of the dataset
```

```python
df.describe()  # Get summary statistics (mean, std, min, max, etc.)
```

```python
# Check for missing values in the dataset
df.isna().sum()
```

```python
print(df.columns)
```

```python
cdf = df[["R&DSpend", "Administration", "MarketingSpend", "State", "Profit"]]
X = cdf[["Administration", "MarketingSpend"]].values
Y = cdf[["Profit"]].values
```

```python
# creating a 1*2 subplot
fig, axs = plt.subplots(1, 2, figsize=(12, 5))

# 1st scatter
axs[0].scatter(cdf.Administration, cdf.Profit, color="blue")
axs[0].set_xlabel("Administration")
axs[0].set_ylabel("Profit")
axs[0].set_title(" ADMINISTRATION VS PROFIT")

# 2nd scatter
axs[1].scatter(cdf.MarketingSpend, cdf.Profit, color="blue")
axs[1].set_xlabel("MarketSpend")
axs[1].set_ylabel("Profit")
axs[1].set_title(" MarketSpend VS PROFIT")

# adjust the layout

plt.tight_layout()
plt.show()
```

### Splitting the train and test dataset

```python
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=4)
```

### Fitting the model

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(x_train, y_train)
```

```python
print(f"coefficient: {lr.coef_}")
print(f"intercept: {lr.intercept_}")
```

```python
y_hat = lr.predict(x_test)
```

```python
from sklearn.metrics import r2_score

print(f"the R2 SCORE is:\n{r2_score(y_test,y_hat)}")
```
